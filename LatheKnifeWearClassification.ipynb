# Training and creation of the classification model

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn import linear_model, metrics
X = df[['Fy', 'Fz', 'F',
       ]]
y = df['TCond']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=123)
```

The Exp2 used the tool condition as a factor, bringing the necessity of having the tool worn to different levels---new tool (VBB = 0.0 mm), mid-life tool (VBB = 0.1 mm), and end-of-life tool (VBB = 0.3 mm). Those values were transformed to respectively 0, 1 and 2. VBB describes tool flank wear. As tool wear builds, increased cutting force may be required to compensate. There will be acceptable tolerance increases for both the tool and the workpiece. But if the wear is significant, the force could exceed acceptable tolerance and require changing.

```python
from sklearn import preprocessing
from sklearn import utils

lab = preprocessing.LabelEncoder()
y_transformed = lab.fit_transform(y_train)
ytest_transformed = lab.fit_transform(y_test)

print(y_transformed)
```

```python
reg = linear_model.LogisticRegression(penalty='l2',C=1, max_iter=800)

reg.fit(X_train, y_transformed)

y_pred = reg.predict(X_test)
```

```python
print("Logistic Regression model accuracy(in %):",
      metrics.accuracy_score(ytest_transformed, y_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(ytest_transformed, y_pred, labels=reg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=reg.classes_)
disp.plot()

plt.show()
```

```python
X_test2 = df2[['Fy', 'Fz', 'F',
       ]]

y_pred2 = reg.predict(X_test2)
y_test2 = df2['TCond']
ytest_transformed2 = lab.fit_transform(y_test2)
print(ytest_transformed2)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


accuracy = accuracy_score(ytest_transformed2, y_pred2)
f1 = f1_score(ytest_transformed2, y_pred2, average='macro')


print("Accuracy:", accuracy)
print("F1-score:", f1)
```

Best results were achieved using model that was trained on passive force, feed force and resultant force.
